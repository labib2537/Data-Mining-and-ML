```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sb
```

```python
from sklearn.model_selection import train_test_split
train = pd.read_csv('gdrive/My Drive/OOP-2 Project/Project Code and CSV file/bank.csv')
train.head(10)
```

```python
train.shape
```

**Data Visualization**

```python
sb.set_style('darkgrid')
sb.set_color_codes(palette='dark')

f, ax = plt.subplots(figsize=(14,9))
sb.countplot(train['job'])
ax.set(title="Histogram for Job")
plt.show()
```

```python
sb.set_style("whitegrid")
plt.figure(figsize=(10,8))
sb.countplot(train['marital'])
plt.show()
```

```python
sb.set_style("whitegrid")
plt.figure(figsize=(10,8))
sb.countplot(train['contact'])
plt.show()
```

```python
sb.set_style("whitegrid")
plt.figure(figsize=(12,9))
sb.countplot(train['education'])
plt.show()
```

```python
plt.figure(figsize=(14,10))
cor = train.corr()
sb.heatmap(cor,annot=True)
plt.show()
```

```python
chart = sb.catplot(data= train, x = 'campaign', y= 'previous', kind= 'box', height=10, palette='Set2')
chart.set_xticklabels(fontweight ='light', fontsize = 'large')
```

```python
plt.figure(figsize=(10,10))

sb.scatterplot(data=train, x='campaign', y='pdays', alpha=0.65, color='r')
```
